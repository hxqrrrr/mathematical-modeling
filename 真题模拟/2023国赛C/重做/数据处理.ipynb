```python
plt.rcParams['font.sans-serif'] = ['SimHei']  # 用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False  # 用来正常显示负号
def seasonal_plot(X, y, period, freq, ax=None, n_ticks=7):
    if ax is None:
        _, ax = plt.subplots()
    palette = sns.color_palette("husl", n_colors=X[period].nunique(),)
    ax = sns.lineplot(
        x=freq,
        y=y,
        hue=period,
        data=X,
        ci=False,
        ax=ax,
        palette=palette,
        legend=False,
    )
    ax.set_title(f"Seasonal Plot ({period}/{freq})")
    
    # 设置 x 轴刻度
    x_values = X[freq].unique()
    step = max(len(x_values) // n_ticks, 1)  # 确保 step 至少为 1
    ax.set_xticks(x_values[::step])
    ax.set_xticklabels(x_values[::step], rotation=45, ha='right')
    
    for line, name in zip(ax.lines, X[period].unique()):
        y_ = line.get_ydata()[-1]
        ax.annotate(
            name,
            xy=(1, y_),
            xytext=(6, 0),
            color=line.get_color(),
            xycoords=ax.get_yaxis_transform(),
            textcoords="offset points",
            size=14,
            va="center",
        )
    return ax



def plot_periodogram(ts, detrend='linear', ax=None):
    from scipy.signal import periodogram
    fs = pd.Timedelta("365D") / pd.Timedelta("1D")
    freqencies, spectrum = periodogram(
        ts,
        fs=fs,
        detrend=detrend,
        window="boxcar",
        scaling='spectrum',
    )
    if ax is None:
        _, ax = plt.subplots()
    ax.step(freqencies, spectrum, color="purple")
    ax.set_xscale("log")
    ax.set_xticks([1, 2, 4, 6, 12, 26, 52, 104])
    ax.set_xticklabels(
        [
            "Annual (1)",
            "Semiannual (2)",
            "Quarterly (4)",
            "Bimonthly (6)",
            "Monthly (12)",
            "Biweekly (26)",
            "Weekly (52)",
            "Semiweekly (104)",
        ],
        rotation=30,
    )
    ax.ticklabel_format(axis="y", style="sci", scilimits=(0, 0))
    ax.set_ylabel("Variance")
    ax.set_title("Periodogram")
    return ax
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
def monthly_daily_plot(X, y, date_column, ax=None):
    if ax is None:
        _, ax = plt.subplots(figsize=(15, 6))
    
    # 确保日期列是datetime类型
    X[date_column] = pd.to_datetime(X[date_column])
    
    # 创建月份和日期列
    X['month'] = X[date_column].dt.month
    X['day'] = X[date_column].dt.day
    
    # 为每个月选择不同的颜色
    palette = sns.color_palette("husl", n_colors=12)
    
    # 绘制线图
    ax = sns.lineplot(
        x='day',
        y=y,
        hue='month',
        data=X,
        ci=False,
        ax=ax,
        palette=palette,
    )
    
    ax.set_title(f"每月每天{y}变化")
    ax.set_xlabel("日")
    ax.set_ylabel(y)
    
    # 设置x轴刻度为30天
    ax.set_xticks(range(1, 31))
    ax.set_xticklabels(range(1, 31))
    
    # 添加图例
    ax.legend(title="月份", bbox_to_anchor=(1.05, 1), loc='upper left')
    
    # 替换图例标签为月份名称
    handles, labels = ax.get_legend_handles_labels()
    ax.legend(handles, [calendar.month_abbr[int(label)] for label in labels], 
              title="月份", bbox_to_anchor=(1.05, 1), loc='upper left')
    
    return ax

def yearly_plot(X, y, year_column, date_column, ax=None):
    if ax is None:
        _, ax = plt.subplots(figsize=(15, 6))
    
    # 确保日期列是datetime类型
    X[date_column] = pd.to_datetime(X[date_column])
    
    # 创建一个表示一年中的天数的列
    X['day_of_year'] = X[date_column].dt.dayofyear
    
    # 为每一年选择不同的颜色
    palette = sns.color_palette("husl", n_colors=X[year_column].nunique())
    
    # 绘制线图
    ax = sns.lineplot(
        x='day_of_year',
        y=y,
        hue=year_column,
        data=X,
        ci=False,
        ax=ax,
        palette=palette,
    )
    
    ax.set_title(f"Yearly Plot ({y})")
    ax.set_xlabel("Day of Year")
    ax.set_ylabel(y)
    
    # 设置x轴刻度为每月的第一天
    month_ticks = [1, 32, 60, 91, 121, 152, 182, 213, 244, 274, 305, 335]
    month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
    ax.set_xticks(month_ticks)
    ax.set_xticklabels(month_names)
    
    # 添加图例
    ax.legend(title="Year", bbox_to_anchor=(1.05, 1), loc='upper left')
    
    return ax

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import calendar



df_merge_groupby = df_merge.groupby(['单品编码', '销售日期','分类编码']) \
    .agg({'销量(千克)': 'sum', '收款额': 'sum', '销售单价(元/千克)': 'mean'}) \
    .reset_index()
df_merge_groupby['day'] =df_merge_groupby['销售日期'].apply(lambda x:x.weekday())
df_merge_groupby['week'] = df_merge_groupby['销售日期'].apply(lambda x:x.month)
df_merge_groupby['季节'] = df_merge_groupby['销售日期'].apply(lambda x:x.quarter)
df_merge_groupby['年']  =   df_merge_groupby['销售日期'].apply(lambda x:x.year)
fig, (ax0, ax1,ax2) = plt.subplots(3, 1, figsize=(16, 6))

seasonal_plot(df_merge_groupby, y="销量(千克)", period="week", freq="day", ax=ax0)
monthly_daily_plot(df_merge_groupby, y="销量(千克)", date_column="销售日期", ax=ax1)
yearly_plot(df_merge_groupby, y="销量(千克)", year_column="年", date_column="销售日期", ax=ax2)
```

```python
from statsmodels.tsa.deterministic import CalendarFourier, DeterministicProcess
from sklearn.linear_model import LinearRegression
import pandas as pd

# 假设 df_merge_groupby 是您之前的数据框架，包含日期索引和"销量(千克)"列




fourier = CalendarFourier(freq="A", order=10)  # 10对正弦/余弦项用于年度季节性

dp = DeterministicProcess(
    index=df_merge_groupby.index,
    constant=True,               # 虚拟特征用于偏差（y截距）
    order=1,                     # 趋势（1阶表示线性）
    seasonal=True,               # 周季节性（指标）
    additional_terms=[fourier],  # 年度季节性（傅里叶）
    drop=True,                   # 删除项以避免共线性
)

X = dp.in_sample()  # 为 df_merge_groupby.index 中的日期创建特征
```

```python

# 现在我们的特征集已创建，准备拟合模型并进行预测
# 我们将添加90天的预测，以查看模型如何在训练数据之外进行外推

y = df_merge_groupby["销量(千克)"]

model = LinearRegression(fit_intercept=False)
_ = model.fit(X, y)

y_pred = pd.Series(model.predict(X), index=y.index)
X_fore = dp.out_of_sample(steps=90)
y_fore = pd.Series(model.predict(X_fore), index=X_fore.index)

ax = y.plot(color='0.25', style='.', title="销量预测 - 季节性预测")
ax = y_pred.plot(ax=ax, label="季节性")
ax = y_fore.plot(ax=ax, label="季节性预测", color='C3')
_ = ax.legend()
df_merge_groupby["销量(千克)"]
```

#####   季节性指标

```python
def get_season(b):
    length = len(b)  # 使用 'length' 避免与内置函数冲突
    if length < 50:
        return 1
    elif length > 500:
        return 3
    else:
        return 2

def get_index(df_merge_groupby):
    a = df_merge_groupby.groupby('单品编码')['销售日期'].count()
    length = a.shape[0]
    result = []
    
    for i in range(length):
        # 获取对应单品编码的数据
        b = df_merge_groupby[df_merge_groupby['单品编码'] == a.index[i]]
        
        # 计算季节并将结果添加到列表中
        result.append(get_season(b))
    
    return result

# 确保在调用 get_index 函数时传入正确的 DataFrame
```
